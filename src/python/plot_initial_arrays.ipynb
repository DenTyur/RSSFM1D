```python
import numpy as np
from matplotlib import cm
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import os
basedir = os.path.abspath(os.getcwd())
src_dir = os.path.abspath(os.path.join(basedir, '..'))
src_dir
```

```python
x = np.load(src_dir + "/arrays_saved/x.npy")
atomic_potential = np.load(src_dir + "/arrays_saved/atomic_potential.npy")
psi_initial = np.load(src_dir + "/arrays_saved/psi_initial.npy")
```

```python
fig, ax = plt.subplots(1, 1, figsize=(5, 5))
ax.set(title = "Atomic and Absorbing potentials")
ax.plot(x, np.abs(atomic_potential), color="black")
```

```python
fig, ax = plt.subplots(1, 1, figsize=(5, 5))
ax.set(title = "Initial Wave Function")
ax.plot(x, np.abs(psi_initial), color="blue")
```
